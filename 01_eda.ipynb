```python
import pandas as pd
df = pd.read_excel('../data/appstoregames.xlsx')
print(df.shape)
df.head()
```

```python
duplicate_count = df.duplicated(subset=['App ID']).sum()
print(f"Số lượng game bị trùng lặp: {duplicate_count}")
```

```python
if duplicate_count > 0:
    df = df.drop_duplicates(subset=['App ID'], keep='first')
    print("Done!")
```

```python
df.isnull().sum()
```

```python
import numpy as np
#1. Xóa cột ko cần thiết
columns_to_drop = ['App URL', 'Icon URL', 'Subtitle']
df = df.drop(columns=columns_to_drop)
```

```python
df.info()
```

```python
#2. Xử lý missing values
df = df.dropna(subset=['Average User Rating', 'User Rating Count'])
```

```python
df = df.dropna(subset=['Name', 'Size in Bytes', 'Description'])
```

```python
df['Size in MB'] = round(df['Size in Bytes'] / (1024 * 1024), 2)
```

```python
df = df.drop(columns=['Size in Bytes'])
```

```python
#Phân loại game miễn phí/ trả phí
df['Is Free'] = np.where(df['Price per App (USD)'] == 0, 'Free', 'Paid')
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:your_password_here@localhost:5432/mobile_game_db')

df.to_sql('mobile_games', engine, if_exists='replace', index=False)
print("Done!")
```
